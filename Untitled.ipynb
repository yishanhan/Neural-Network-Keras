```python
import pandas as pd
import numpy as np
import re
```

```python
# Import dataset 
dataset = pd.read_csv("DeciBio Sample.csv")                     
df = pd.DataFrame(dataset)

# Separate names and addresses 
# By finding all the list of names in [] so we end up with [name; name; name]
# For address, find all that is separated by , and ] 
names = df.AuthorAddress.apply(lambda x : re.findall('\[[^\]]*]', x))
address = df.AuthorAddress.apply(lambda x : re.findall('\,[^[]*', x))

names
```

```python
# Create empty lists so we do not overwrite actual dataset
# Go through each list, and within each list, count how many names there are so we can know
# how many names are associated with each address
# Split by ; after we are done 
names_per_address = []
counts_per_address = []
for each_row in range(len(names)):
    for each_name in range(len(names[each_row])):
        names_per_address.append(names[each_row][each_name].split(';'))
        counts_per_address.append(len(names[each_row][each_name].split(';')))


names_per_address
```

```python
# Take out the individual names
names_general = []
for i in range(len(names_per_address)):
    for j in range(len(names_per_address[i])):
        names_general.append(names_per_address[i][j])

name_country = []
for each_row in range(len(address)):
    for each_address in range(len(address[each_row])):
        name_country.append(address[each_row][each_address])

for i in range(len(name_country)):
    name_country[i] = name_country[i][-10:]
    name_country[i] = str(name_country[i]).split(' ', 1)[1]
    
names_general
```

```python
country_full = []
for i in range(len(counts_per_address)):
    country_full.append(name_country[i] * counts_per_address[i])
    
country_general = []
for i in range(len(country_full)):
    country_general.append(country_full[i])

country_full
```

```python
d = {'Author' : np.array(names_per_address),
    "Address" : np.array(country_full)}
```

```python
df = pd.DataFrame(data = d)
```
